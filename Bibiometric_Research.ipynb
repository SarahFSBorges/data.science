## Generate bibliometric graphics

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
# read text from file
with open("df_final.csv", "r") as file:
    text = file.read()

# create wordcloud
wordcloud = WordCloud().generate(text)

# plot wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("on")
plt.show()
```

### Updating wordcloud parameters

```python
# read text from file
stopwords = set("df_final.csv")

# create stopwords list
stopwords = set(STOPWORDS)
stopwords.update(["in", "the", "of", "on", "M", "The", "and","S", "J","E","D","C","G","P","V","Citation","Index"])

# create wordcloud
wordcloud = WordCloud(width=800, height=400, 
                      stopwords=stopwords, 
                      background_color="#FFFFCC", 
                      min_font_size=10).generate(text)

# plot wordcloud
plt.figure(figsize=(10, 7), dpi=100)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

plt.imshow(wordcloud);
wordcloud.to_file("final_file.png")
```

```python
!pip install pybliometrics
from pybliometrics.scopus import ScopusSearch

# Define the search query and perform the search
query = ScopusSearch('Year > 2019', count=1000)

# Create a dictionary to store the country counts
country_counts = {}

# Iterate through the results and count the number of publications per country
for result in query.results:
    country = result['affiliation'][0]['affilname']
    if country in country_counts:
        country_counts[country] += 1
    else:
        country_counts[country] = 1

# Print the top 10 countries by number of publications
for country, count in sorted(country_counts.items(), key=lambda item: item[1], reverse=True)[:10]:
    print(f"{country}: {count} publications")
```
