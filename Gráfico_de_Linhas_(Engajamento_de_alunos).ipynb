# Verifique a taxa de engajamento dos alunos (nota individual x nota do grupo do projeto)

```python
# importando pacotes
import matplotlib.pyplot as plt
import numpy as np
```

```python
# configurar a fonte
plt.rcParams['font.family'] = 'Arial'

# configurar as cores que usarei no plot
AZUL = "#003F5c" 
CINZA = "#9E9E9E"
CINZA_TEXTO = "#515151"
```

```python
# dados ficticios 
X = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
alunos = [70, 90, 70, 80, 80, 90, 70, 80, 80, 90, 80, 100]
projetos = [70, 60, 50, 60, 60, 70, 60, 60, 50, 60, 60, 70]
```

```python
print("\n\n\n\n")

# definindo a figura
fig, ax = plt.subplots(figsize=(10,6), dpi=72, )

# plots das linhas
ax.plot(range(12), alunos, color=CINZA, linewidth=4)
ax.plot(range(12), projetos, color=AZUL, linewidth=4)

# plots dos scatter plots
ax.scatter(range(12)[-3:], alunos[-3:], s=90, color=CINZA, clip_on=False)
ax.scatter(range(12)[-3:], projetos[-3:], s=90, color=AZUL, clip_on=False)

# anotações nos pontos de scatter
for i, j in enumerate(alunos[-3:]):
    ax.annotate(
        j,
        xy=(i+9, j),
        xytext=(5, 5),
        textcoords='offset points',
        color=CINZA_TEXTO,
        fontsize=12,
        fontweight='medium'
    )

for i, j in enumerate(projetos[-3:]):
    ax.annotate(
        j,
        xy=(i+9, j),
        xytext=(5, 5),
        textcoords='offset points',
        color=AZUL,
        fontsize=12,
        fontweight='medium'
    )

# legenda no final de cada linha
ax.annotate(
    "Alunos",
    xy=(len(alunos)-1, alunos[-1]),
    xytext=(50,0),
    textcoords='offset points',
    color=CINZA,
    fontsize=18,
    fontweight='semibold'
            )

ax.annotate(
    "Projetos",
    xy=(len(projetos)-1, projetos[-1]),
    xytext=(50,0),
    textcoords='offset points',
    color=AZUL,
    fontsize=18,
    fontweight='semibold'
            )

# remover spines da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar range do eixo y
ax.set_ylim(0, 100)

# remover yticks e yticklabels
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# configurar xticks (colocar tocos ticks de 12 meses)
ax.set_xticks(ticks=range(12))
ax.set_xticklabels(labels=X)

# title
ax.text(-0.5, 120,'Taxa de Engajamento', fontsize=24, color=CINZA_TEXTO, fontweight='bold')
ax.text(-0.5, 100,'KPI de engajamento, que considera a quantidade\nde acessos únicos e projetos enviados para\nrevisão  do Sigmoidal', fontsize=13, color=CINZA)

# rodapé
ax.text(
    -0.5,
    -15,
    "Fonte: Dados extraídos a partir da Hotmart e Memberkit (21/06/2021)",
    fontsize=12, color=CINZA
    )

# exibir grafico
plt.show()
```
