# Visualizar e estudar o mapa de Dublin.

```python
# importar biblitecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
# importar dados
df = pd.read_csv("http://data.insideairbnb.com/ireland/leinster/dublin/2022-09-11/data/listings.csv.gz")
```

```python
# verificar tamanho do dataset
print('Dataset original: ', df.shape)

# criar dataset menor (10%)
df_map = df.sample(frac=0.01, random_state=0)

# verificar novo tamanho do dataset
print('Tamanho do Dataset: ', df_map.shape)
```

```python
# criar variaveis
latitude = df_map.latitude.values
longitude = df_map.longitude.values
listings = df_map.name.values
```

```python
# criar mapa
dublin = folium.Map(location=[53.350140,-6.266155],
                    zoom_start=10)
dublin
```

### Pontos de hospedagem em Dublin.

```python
# adicionar marcadores
for lat, lon, listing in zip(latitude, longitude, listings):
  folium.Marker(location=[float(lat), float(lon)], popup=listing).add_to(dublin)

# plotar mapa
dublin.save('dublin_map.html')
dublin
```

```python
# instalar PyDeck para gráficos interativos
!pip install pydeck -q

import pydeck as pdk

attractions = 'https://failteireland.azure-api.net/opendata-api/v1/attractions/csv' 

layer = pdk.Layer(
    'HexagonLayer',  
    attractions,
    get_position=['longitudeg', 'latitude'],
    auto_highlight=True,
    elevation_scale=50,
    pickable=True,
    elevation_range=[0, 3000],
    extruded=True,
    coverage=1)

# Localização de Dublin
view_state = pdk.ViewState(
    longitude=-6.266155,
    latitude=53.350140,
    zoom=10,
    min_zoom=4,
    max_zoom=15,
    pitch=40.5,
    bearing=-27.36)

# Combined all of it and render a viewport
r = pdk.Deck(layers=[layer], initial_view_state=view_state)
r.to_html('hexagon-example.html')
```
