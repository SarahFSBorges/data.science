# Terminologia

### Instância: o elemento que receberá uma previsão. Por exemplo, a instância pode ser uma página da Web que você quer classificar como "sobre gatos" ou "não sobre gatos".
### Identificador: uma resposta para uma tarefa de previsão, ou seja, a resposta produzida por um sistema de machine learning ou a resposta correta fornecida nos dados de treinamento. Por exemplo, o rótulo de uma página da Web pode ser "quot;about cats"" ou "gatos".
### Recurso: uma propriedade de uma instância usada em uma tarefa de previsão. Por exemplo, uma página da Web pode ter um recurso "contém a palavra 'cat'".
### Coluna de recursos: um conjunto de recursos relacionados, como o conjunto de todos os países possíveis em que os usuários podem viver. Um exemplo pode ter um ou mais atributos presentes em uma coluna de atributo. "Coluna de atributo" é uma terminologia específica do Google. Uma coluna de atributo é chamada de "namespace" no sistema VW (no Yahoo/Microsoft) ou em um campo.
### Exemplo: uma instância (com os recursos) e um rótulo.
### Modelo: é uma representação estatística de uma tarefa de previsão. Você treina um modelo em exemplos e o usa para fazer previsões.
### Métrica: um número importante para você. Pode ser otimizada diretamente ou não.
### Objetivo: é uma métrica que seu algoritmo está tentando otimizar.
### Pipeline: a infraestrutura ao redor de um algoritmo de machine learning. Inclui coletar os dados do front-end, colocá-los em arquivos de dados de treinamento, treinar um ou mais modelos e exportar os modelos para produção.
### Taxa de cliques: a porcentagem de visitantes de uma página da Web que clicam em um link de um anúncio.
